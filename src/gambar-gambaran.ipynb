```python
import pandas as pd
import matplotlib.pyplot as plt

# Set the path to the CSV file
csv_file_path = "crema-d-adam_0001.csv"

# Load the CSV data using Pandas
df = pd.read_csv(csv_file_path)

# Extract the desired columns for plotting
epoch = df["epoch"]
loss = df["val_loss"]
accuracy = df["val_accuracy"]

# Plot the loss and accuracy curves
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(epoch, loss)
plt.title("Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.subplot(1, 2, 2)
plt.plot(epoch, accuracy)
plt.title("Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")

plt.tight_layout()
plt.show()
```

```python
# Smoothed Version
# Set the path to the CSV file
csv_file_path = "SAVEElog/60epok/SAVEE_CNN_LSTM_adam_00001.csv"

# Load the CSV data using Pandas
df = pd.read_csv(csv_file_path)

# Extract the desired columns for plotting
epoch = df["epoch"]
loss = df["val_loss"]
accuracy = df["val_accuracy"]

# Set the window size for the moving average
window_size = 1

# Calculate the moving averages
loss_smooth = loss.rolling(window_size, min_periods=1).mean()
accuracy_smooth = accuracy.rolling(window_size, min_periods=1).mean()

# Plot the smoothed loss and accuracy curves
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(epoch, loss_smooth)
plt.title("Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.subplot(1, 2, 2)
plt.plot(epoch, accuracy_smooth)
plt.title("Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")

plt.tight_layout()
plt.show()
```

```python
n_mfcc = 40
hop_length = 512
output_dir = "GambarGambar"
for folder in folders:
    folder_path = os.path.join(inspect_path, folder)
    
    # Iterate over the audio files in the folder
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        
        # Load the audio file
        audio, sr = librosa.load(file_path)
        
        # Compute MFCC features
        mfcc = librosa.feature.mfcc(audio, sr=sr, n_mfcc=n_mfcc, hop_length=hop_length)

        # Plot the MFCC features
        plt.figure(figsize=(4, 2))
        librosa.display.specshow(mfcc, x_axis='time')
        plt.colorbar(format='%+2.0f dB')
        plt.title(file_name)
        plt.xlabel('Time')
        plt.ylabel('MFCC Coefficients')
        plt.tight_layout()
        # Save the plots
        output_file_name = os.path.splitext(file_name)[0]  # Remove file extension
        output_file_path = os.path.join(output_dir, output_file_name)
        plt.savefig(output_file_path + '.png')
        plt.close()
        plt.show()
        # # Save the plots
        # output_file_name = os.path.splitext(file_name)[0]  # Remove file extension
        # output_file_path = os.path.join(output_dir, output_file_name)
        # plt.savefig(output_file_path + '.png')
        # plt.close()
```

```python
mfcc_pitch_shifted = librosa.effects.pitch_shift(mfcc, sr, n_steps=2)  # Example: Pitch-shifted MFCCs
mfcc_noisy = mfcc + np.random.normal(0, 1, size=mfcc.shape)  # Example: Noisy MFCCs
mfcc_time_stretched = librosa.effects.time_stretch(mfcc, rate=0.8)  # Example: Time-stretched MFCCs

mfcc_pitch_shifted_diff = mfcc_pitch_shifted - mfcc
mfcc_noisy_diff = mfcc_noisy - mfcc
mfcc_time_stretched_diff = mfcc_time_stretched - mfcc

plt.figure(figsize=(12, 8))

# Plot original MFCCs
plt.subplot(2, 2, 1)
librosa.display.specshow(mfcc, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.xlabel('Time')
plt.ylabel('MFCC Coefficients')
plt.title('Original')

# Plot pitch-shifted differences
plt.subplot(2, 2, 2)
librosa.display.specshow(mfcc_pitch_shifted_diff, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.xlabel('Time')
plt.ylabel('MFCC Coefficients')
plt.title('Pitch Shifted Difference')

# Plot noisy differences
plt.subplot(2, 2, 3)
librosa.display.specshow(mfcc_noisy_diff, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.xlabel('Time')
plt.ylabel('MFCC Coefficients')
plt.title('Noisy Difference')

# Plot time-stretched differences
plt.subplot(2, 2, 4)
librosa.display.specshow(mfcc_time_stretched_diff, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.xlabel('Time')
plt.ylabel('MFCC Coefficients')
plt.title('Time Stretched Difference')

plt.tight_layout()
plt.show()
```

```python
n_mfcc = 40
hop_length = 512
output_dir = "GambarGambar"
# RAVDESS
n_mfcc = 40
hop_length = 512
output_dir = "GambarGambar"

AugmentedGN = "dataset/Augmented/AddGaussianNoise/1001_DFA_DIS_XX.wav"
AugmentedPS = "dataset/Augmented/PitchShift/1001_DFA_DIS_XX.wav"
AugmentedTS = "dataset/Augmented/TimeStretch/1001_DFA_DIS_XX.wav"
Original = "dataset/Augmented/original/1001_DFA_DIS_XX.wav"
x = [AugmentedGN, AugmentedPS, AugmentedTS, Original]


for file in x:
    audio, sr = librosa.load(file)
    mfcc = librosa.feature.mfcc(audio, sr=sr, n_mfcc=n_mfcc, hop_length=hop_length)

    # Generate modified versions of MFCCs
    mfcc_pitch_shifted = librosa.effects.pitch_shift(mfcc, sr, n_steps=2)
    mfcc_noisy = mfcc + np.random.normal(0, 1, size=mfcc.shape)
    mfcc_time_stretched = librosa.effects.time_stretch(mfcc, rate=0.8)

    # Resize the modified MFCCs to match the shape of the original MFCCs
    mfcc_pitch_shifted_resized = librosa.util.fix_length(mfcc_pitch_shifted, mfcc.shape[1])
    mfcc_noisy_resized = librosa.util.fix_length(mfcc_noisy, mfcc.shape[1])
    mfcc_time_stretched_resized = librosa.util.fix_length(mfcc_time_stretched, mfcc.shape[1])

    # Calculate the differences between the modified MFCCs and the original MFCCs
    mfcc_pitch_shifted_diff = mfcc_pitch_shifted_resized - mfcc
    mfcc_noisy_diff = mfcc_noisy_resized - mfcc
    mfcc_time_stretched_diff = mfcc_time_stretched_resized - mfcc

    # Plot the original and modified MFCCs, along with their differences
plt.figure(figsize=(12, 8))

# Plot original MFCCs
plt.subplot(2, 2, 1)
librosa.display.specshow(mfcc, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.xlabel('Time')
plt.ylabel('MFCC Coefficients')
plt.title('Original')

# Plot pitch-shifted differences
plt.subplot(2, 2, 2)
librosa.display.specshow(mfcc_pitch_shifted_diff, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.xlabel('Time')
plt.ylabel('MFCC Coefficients')
plt.title('Original - Pitch Shifted')

# Plot noisy differences
plt.subplot(2, 2, 3)
librosa.display.specshow(mfcc_noisy_diff, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.xlabel('Time')
plt.ylabel('MFCC Coefficients')
plt.title('Original - Noisy')

# Plot time-stretched differences
plt.subplot(2, 2, 4)
librosa.display.specshow(mfcc_time_stretched_diff, x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.xlabel('Time')
plt.ylabel('MFCC Coefficients')
plt.title('Original - Time Stretched')

plt.tight_layout()
plt.show()
```
