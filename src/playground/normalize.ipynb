```python
import os
import pandas as pd
import numpy as np
import librosa
import math
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
```

```python
train_data_value = np.load('../saved_dataset/40_2048_512_train_data_value.npy')
train_data_target = np.load('../saved_dataset/40_2048_512_train_data_target.npy')
test_data_value = np.load('../saved_dataset/40_2048_512_test_data_value.npy')
test_data_target = np.load('../saved_dataset/40_2048_512_test_data_target.npy')
```

```python
combined_data = np.concatenate((train_data_value, test_data_value), axis=0)
combined_data = np.reshape(combined_data, (7442, -1))

```

```python

scaler = StandardScaler()
scaler.fit(combined_data)
# scaled_data_2d = scaler.transform(data_2d)
# scaled_data = np.reshape(scaled_data_2d, (5953, 156, 40))
```

```python
train_data_value.shape[1]
```

```python
train_data_value_2d = np.reshape(train_data_value, (train_data_value.shape[0], -1))
test_data_value_2d = np.reshape(test_data_value, (test_data_value.shape[0], -1))
normalized_train_value = scaler.transform(train_data_value_2d)
normalized_test_value = scaler.transform(test_data_value_2d)
```

```python
normalized_train_value = np.reshape(normalized_train_value, (train_data_value.shape[0], train_data_value.shape[1], train_data_value.shape[2]))
normalized_test_value = np.reshape(normalized_test_value, (test_data_value.shape[0], test_data_value.shape[1], test_data_value.shape[2]))
```

```python
np.save("../saved_dataset/40_2048_512_train_data_value_normalized.npy", normalized_train_value)
np.save("../saved_dataset/40_2048_512_test_data_value_normalized.npy", normalized_test_value)
```
