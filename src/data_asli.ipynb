```python
import numpy as np
import librosa
import os
import matplotlib.pyplot as plt
import librosa.display
```

```python
folder_path = "DataIndo"
```

```python
for x in os.listdir(folder_path):
    file_path = os.path.join(folder_path, x)
    signal, sr = librosa.load(file_path)
    features = librosa.feature.mfcc(y=signal, sr=48000, n_mfcc=40, n_fft=2048, hop_length=512)
    plt.figure(figsize=(4, 2))
    librosa.display.specshow(features, x_axis='time')
    plt.colorbar(format='%+2.0f dB')
    # plt.title(x)
    plt.xlabel('Time')
    plt.ylabel('MFCC Coefficients')
    plt.tight_layout()
    plt.savefig("GambarGambar/gambar2/{}.jpg".format(x[:-4]))
    plt.show()
```
